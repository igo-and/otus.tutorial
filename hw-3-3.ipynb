#### Найдите количество курящих людей

```python
tips.value_counts('smoker')
```

```python
tips[tips['smoker'] == 'No'].count()
```

#### Узнайте какой средний 'total_bill' в зависимости от 'day'
#### а также минимальный и максимальный чек и медиану

```python
tips.groupby('day').total_bill.agg(['mean', 'min', 'max', 'median'])
```

#### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
med = float(tips.total_bill.agg(['median']))

med
```

```python
tips_more = tips[tips['total_bill'] > med ] 
tips_more
```

##### какой средний 'tip' в зависимости от 'sex'
##### а так же мин и иакс чаевые

```python
tips_more.groupby('sex').tip.agg(['mean', 'min', 'max'])
```

#### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips_binary = tips.copy()
```

```python
map_dict = {'No' : 0,
           'Yes' : 1}
```

```python
tips_binary['smoker'] = tips_binary['smoker'].map(map_dict)
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'
